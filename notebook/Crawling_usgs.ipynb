Import Library

```python
import requests
import pandas as pd
```

Test API USGS

```python
url = "https://earthquake.usgs.gov/fdsnws/event/1/query"

params = {
    "format": "geojson",
    "starttime": "2025-01-01",
    "endtime": "2025-01-31"
}

response = requests.get(url, params=params)

print(response.status_code)
```

Ambil Data dalam Format JSON

```python
data = response.json()

len(data["features"])
```

Melihat Struktur Data

```python
data["features"][0]
```

Mengambil Data Penting

```python
quake = data["features"][0]
quake

quake["properties"]["mag"]
quake["properties"]["place"]
```

Mengubah JSON menjadi Tabel

```python
records = []

for item in data["features"]:

    properties = item["properties"]
    geometry = item["geometry"]

    records.append({
        "id": item["id"],
        "magnitude": properties["mag"],
        "place": properties["place"],
        "time": properties["time"],
        "updated": properties["updated"],
        "longitude": geometry["coordinates"][0],
        "latitude": geometry["coordinates"][1],
        "depth": geometry["coordinates"][2]
    })

print(len(records))
```

Membuat DataFrame

```python
df = pd.DataFrame(records)

df.head()
```

Cek Struktur Data

```python
df.shape
df.info()
```

Mengambil Data (2025)

```python
all_records = []

for month in range(1, 13):

    start_date = f"2025-{month:02d}-01"

    if month == 12:
        end_date = "2025-12-31"
    else:
        end_date = f"2025-{month+1:02d}-01"

    print(f"Mengambil {start_date} s/d {end_date}")

    url = "https://earthquake.usgs.gov/fdsnws/event/1/query"

    params = {
        "format": "geojson",
        "starttime": start_date,
        "endtime": end_date
    }

    response = requests.get(url, params=params)

    data = response.json()

    for item in data["features"]:

        properties = item["properties"]
        geometry = item["geometry"]

        all_records.append({
            "id": item["id"],
            "magnitude": properties["mag"],
            "place": properties["place"],
            "time": properties["time"],
            "updated": properties["updated"],
            "longitude": geometry["coordinates"][0],
            "latitude": geometry["coordinates"][1],
            "depth": geometry["coordinates"][2]
        })

print("TOTAL:", len(all_records))
```

Membuat DataFrame Besar

```python
df_big = pd.DataFrame(all_records)
```

Verifikasi Big Data

```python
df_big.shape
df_big.head()
df_big.info()
```

Menyimpan dalam format CSV

```python
df_big.to_csv(
    "../data/usgs_2025.csv",
    index=False
)
```
