Tes Koneksi Python ke SQL Server

```python
from sqlalchemy import create_engine

server = "HERU"

database = "USGS_BIGDATA"

engine = create_engine(
    f"mssql+pyodbc://{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server&trusted_connection=yes"
)

print("Koneksi berhasil")
```

Load CSV

```python
import pandas as pd

df = pd.read_csv(
    "../data/usgs_2025_final.csv"
)

#Verifikasi Dataset
df.shape

#Membersihkan Nama Kolom
df.columns
```

Migrasi ke SSMS SQL Server

```python
df.to_sql(
    name="earthquake_2025",
    con=engine,
    if_exists="replace",
    index=False,
    chunksize=5000
)

df.shape
```
